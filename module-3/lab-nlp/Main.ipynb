```python
import re
def clean_up(s):
    s =re.sub(r'http\S+',' ', s)
    s =re.sub(r'\\W', " ", s)
    s= re.sub( '[^a-zA-Z]', ' ', s)
    return s.lower()
```

```python
a= "@Ironhack's-#Q website 776-is http://ironhack.com [(2018)])"
```

```python
a=clean_up(a)
```

```python
def tokenize(s):
    lst = word_tokenize(s)
    return lst
```

```python
lemmatizer = WordNetLemmatizer()
porter = PorterStemmer()
def stem_and_lemmatize(lista):
    c_lista = []
    for w in lista:
        w = porter.stem(w)
        w = lemmatizer.lemmatize(w)
        c_lista.append(w)
    return c_lista
```

```python
a=tokenize(a)
```

```python
a=stem_and_lemmatize(a)
```

```python
import nltk
from nltk.corpus import stopwords
b=set(stopwords.words('english'))
b
```

```python
def remove_stopwords(lst):
    l = []
    for w in lst:
        if w not in b:
            l.append(w)
    return l
```
