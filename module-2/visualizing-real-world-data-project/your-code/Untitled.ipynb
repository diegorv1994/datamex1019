```python
import pandas as pd
import requests
import numpy as np
import folium as f
from pymongo import MongoClient
from pandas.io.json import json_normalize
import matplotlib.pyplot as plt
```

```python
client=MongoClient()
db=client.db_companies
colec=db.companies
```

```python
al_menos_1_ofi=list(colec.find({'offices':{'$not':{'$size':0}}},
                                 {'name':1, 'offices':1, '_id':0}))
```

```python
df=pd.DataFrame(al_menos_1_ofi)
```

```python
def get_first(data):
    res=[]
    ofi=[]
    data=data['offices']
    for e in data:
        principal=None   #solo las tienen geodata
        if e[0]['latitude'] and e[0]['longitude']:
            principal={'type':'Point',
                       'coordinates':[
                           e[0]['longitude'],
                           e[0]['latitude']
                       ]}
        ofi.append(principal)
        res.append({
            'total_offices':len(e),
            'lat':e[0]['latitude'],
            'lng':e[0]['longitude'],
            'oficina_principal':principal
        }) 
    return res, ofi
```

```python
first_office=json_normalize(get_first(df)[0])
first_office['oficina_principal']=get_first(df)[1]
first_office.fillna('o', inplace = True)
first_office
```

```python
mapa=f.Map(location=[33,-84], tiles='Stamen Toner', zoom_start=5)
```

```python
stra=['o']
for e in first_office.values:
    if e[0] not in stra:
        f.CircleMarker([e[0], e[1]], radius=10, icon=f.Icon()).add_to(mapa)
```

```python
emp=list(colec.find({'$and':[{'category_code':'software'},{'number_of_employees':{'$gt':500}},{'founded_year':{'$gt':2006}},{'total_money_raised':{'$gt':'$10.00M'}},{'offices':{'$not':{'$size':0}}}]},{'name':1, 'offices':1, '_id':0}))
#Los criterios de selección son los siguientes:
#1. Que sea una compañia de sofware
#2. Un número de empleados mayor a 500
#3. Que se haya fundado después del 2006
#4. Que haya recaudado más de 10 millones
#5. Que tenga oficinas
emp=pd.DataFrame(emp)
emp
```
